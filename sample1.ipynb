```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generate some sample data
data = np.random.randn(100)

# Create a histogram
plt.hist(data, bins=10)  # Adjust bins as needed
plt.xlabel('Data Values')
plt.ylabel('Frequency')
plt.title('Histogram of Data')
plt.show()

# Create a box plot
sns.boxplot(data=data)
plt.xlabel('Data')
plt.title('Box Plot of Data')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(100000)
plt.hist(data, bins=30, edgecolor='black')
plt.title("Histogram of Random Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randn(1000)
plt.hist(data, bins=100, alpha=0.5, label='10 bins', color='maroon')
plt.hist(data, bins=500, alpha=0.5, label='50 bins', color='black')
plt.legend()
plt.title("Effect of Bin Size on Histogram")
```
